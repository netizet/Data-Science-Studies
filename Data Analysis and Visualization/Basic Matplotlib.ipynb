```python
plt.figure(figsize=(12,5))
plt.plot(x, y, 'r')
plt.xlabel('X axis title here')
plt.xticks([1,2,3,4,5],['A','B','C','D','E'])
plt.ylabel('y axis title here')
plt.title('String title here')
plt.show()
```

```python
plt.plot(x,y, 'r--')
plt.plot(y,x, 'g*')
```

```python
# Creating Multiplot on Same Canvas
plt.figure(figsize = (12,8))
plt.subplot(2,2,1)
plt.plot(x, y, 'r--')
plt.title('Ini subplot 1')

plt.subplot(2,2,2)
plt.plot(y, x, 'g*-')
plt.title('ini subplot 2')

plt.subplot(2,2,3)
plt.plot(x,y)
```

```python
# Create figure (empty canvas)
fig = plt.figure()

# Add set of axes to figure
# left, bottom, width, height (range 0, 1)
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

# Plot on that sey of axes
axes.plot(x, y, 'b')
axes.set_xlabel('Set X label')
axes.set_ylabel('Set Y label')
axes.set_title('Set title')
plt.show()
```

```python
# Create figure (empty canvas)
fig = plt.figure(facecolor='pink')

# Add set of axes to figure
# left, bottom, width, height (range 0, 1)
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

# Plot on that sey of axes
axes.plot(x, y, 'b')
axes.set_xlabel('Set X label')
axes.set_ylabel('Set Y label')
axes.set_title('Set title')

axes2.plot(y, x, 'r')
axes2.set_xlabel('Set X2 label')
axes2.set_ylabel('Set Y2 label')
axes2.set_title('Set title2')
axes2.set_axis_off()
```

```python
fig, axes = plt.subplots()

# Now use the axes object to add stuff to the plot
axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from yahoo_historical import Fetcher
dfINDF = Fetcher('INDF.JK', [2011, 1, 1]).getHistorical()
dfINDF = dfINDF.fillna(method = 'ffill')
dfINDF['Date'] = pd.to_datetime(dfINDF['Date'])
dfINDF.set_index ('Date',inplace = True)

fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (20,15))
color = [['r', 'g'], ['b', 'y']]
title = [['Open', 'High'], ['Volume', 'Adj Close']]

for i, j in list(zip(axes, range(1, len(axes)+1))):
    for k, l in list(zip(i, range(1, len(axes[j-1])+1))):
        k.plot(dfINDF[title[j-1][l-1]], color[j-1][l-1])
        k.set_xlabel('Year', fontsize =20)
        k.tick_params(axis = 'both', labelsize = 16)
        k.set_ylabel('Rupiah', fontsize =20)
        k.set_title("INDF {}".format(title[j-1][l-1]), fontsize = 20)
plt.tight_layout(h_pad = 5, w_pad = 5)
```

```python
plt.hist(dfINDF['Open'])
plt.axvline(np.mean(dfINDF['Open']), color='red')
```

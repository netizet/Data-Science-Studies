```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('~/Downloads/anime.csv')
df.head()
```

```python
rating = pd.read_csv('~/Downloads/rating.csv')
rating.head()
```

```python
data = pd.merge(df, rating, on = 'anime_id')
data.head()
```

Karena pada DataFrame df sudah ada kolom rating, maka kolom rating yang awalnya ada di df berubah menjadi rating_x, kolom rating yang berasal dari DataFrame rating berubah menjadi rating_y

```python
ratings = pd.DataFrame(data.groupby('name')['rating_y'].mean())
ratings['ratings count'] = pd.DataFrame(data.groupby('name')['rating_y'].count())
ratings
```

```python
mat = data.pivot_table(index = 'user_id', columns = 'name', values = 'rating_y')
mat.head()
```

mat dibuat agar memudahkan kita untuk mendapatkan data user rating dan judul animenya

```python
Naruto_rating = mat['Naruto']
Naruto_rating
```

```python
similar_Naruto = mat.corrwith(Naruto_rating)
```

Hasil didapatkan dengan menggunakan mat.corrwith()

```python
corr_Naruto = pd.DataFrame(similar_Naruto, columns = ['Correlation'])
corr_Naruto.dropna(inplace = True)
corr_Naruto['ratings count'] = ratings['ratings count']
corr_Naruto[corr_Naruto['ratings count'] > 1500].sort_values('Correlation', ascending = False).head(20)
```

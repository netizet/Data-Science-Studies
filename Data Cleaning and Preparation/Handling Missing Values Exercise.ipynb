Pada tugas kali ini, ada 62 data kosong pada feature Car (kapasitas mobil di garansi rumah).
Isi data kosong pada feature Car dengan mempertimbangkan feature Price (harga rumah) dan/atau Landsize (luas tanah)

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import math
```

```python
melb = pd.read_csv('~/Downloads/melb_data.csv')
```

```python
melb.head()
```

membuat tabel yang menampilkan value Car yg NaN

```python
nullcar = []
melb_2 = melb[['Car', 'Price', 'Landsize']]
for i in range(0,len(melb_2['Car'])):
    if math.isnan(melb_2['Car'][i]) == True:
        nullcar.append(i)
melb_2.iloc[nullcar]
```

Membuat kolom baru, yaitu perbandingan Car dengan Price dan Landsize

```python
melb_2['Car/Price'] = melb_2['Car'] / melb_2['Price']
melb_2['Car/Landsize'] = melb_2['Car'] / melb_2['Landsize']
```

```python
melb_2[melb_2['Landsize'] ==0]
```

```python
plt.figure(figsize=(50,30))
sns.set(font_scale=4)
sns.distplot(melb_2['Car/Price'].dropna())
```

```python
plt.figure(figsize=(50,30))
sns.set(font_scale=4)
```

```python
melb_2[melb_2['Landsize'] ==0]
```

Ternyata pada kolom Landsize banyak yang valuenya 0

```python
melb_2['Car/Price'].describe()
```

```python
melb_2['Car/Landsize'].describe()
```

```python
Q2_cp = np.quantile(melb_2['Car/Price'].dropna(), .50)
Q1_cp = np.quantile(melb_2['Car/Price'].dropna(), .25)
Q3_cp = np.quantile(melb_2['Car/Price'].dropna(), .75)

IQR_cp = (Q3_cp - Q1_cp) * 1.5
batas_bawah_cp = Q1_cp - IQR_cp
batas_atas_cp = Q3_cp + IQR_cp
batas_atas_cp
```

```python
Q2_cl = np.quantile(melb_2['Car/Landsize'].dropna(), .50)
Q1_cl = np.quantile(melb_2['Car/Landsize'].dropna(), .25)
Q3_cl = np.quantile(melb_2['Car/Landsize'].dropna(), .75)

IQR_cl = (Q3_cl - Q1_cl) * 1.5
batas_bawah_cl = Q1_cl - IQR_cl
batas_atas_cl = Q3_cl + IQR_cl
batas_atas_cl
```

```python
c_p = melb_2[(melb_2['Car/Price']>batas_bawah_cp) | (melb_2['Car/Price']<batas_atas_cp)]['Car/Price']
```

```python
c_l = melb_2[(melb_2['Car/Landsize']>batas_bawah_cl) | (melb_2['Car/Landsize']<batas_atas_cl)]['Car/Landsize']
```

```python
with pd.option_context('mode.use_inf_as_na', True):
    c_l = c_l.dropna()
c_l
```

```python
plt.figure(figsize=(50,30))
sns.set(font_scale=4)
sns.distplot(c_p)
```

```python
plt.figure(figsize=(50,30))
sns.set(font_scale=4)
sns.distplot(c_l)
```

Jika diambil rata-ratanya, untuk satu mobil perlu 500,000 AUD atau 207.641196 m^2

```python
c_l.mean()
```

```python
for i in nullcar:
    melb['Car'][i] = (math.floor(melb['Price'][i]*c_p.mean()))
    melb_2['Car'][i]=(math.floor(melb_2['Landsize'][i]*c_l.mean()))
    
# Jadi coba kita bandingkan pengisian dengan pengaruh Price di tabel melb dan pengaruh Landsize di tabel melb_2
```

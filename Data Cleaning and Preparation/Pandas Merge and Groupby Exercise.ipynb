```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
mpg = sns.load_dataset('mpg')
```

## Soal 1

```python
sum(mpg['horsepower'].isna())
```

```python
na_dict={}
for i in mpg.columns:
    na_dict[i] = sum(mpg[i].isna())
pd.Series(na_dict, name='Null Value Count')
```

```python
mpg.isna().sum()
```

```python
horse_na = mpg[mpg['horsepower'].isna()].index
```

```python
mpg.groupby('origin').mean()
```

```python
origin_mean = mpg.groupby('origin').mean()['horsepower']
```

```python
def fillna_origin(cols):
    if pd.isna(cols['horsepower']):
        return round(origin_mean[cols['origin']])
    else:
        return cols['horsepower']
```

```python
mpg[['horsepower', 'origin']]
```

```python
mpg['horsepower'] = mpg[['horsepower', 'origin']].apply(fillna_origin, axis = 1)
```

```python
mpg.loc[horse_na]
```

## Soal 3

```python
mpg['origin'].unique()
```

```python
def add_horse_origin(x):
    if x['horsepower'] <= 105:
        return 'lower {}'.format(x['origin'])
    else:
        return 'upper {}'.format(x['origin'])
    
```

```python
mpg['horse_origin'] = mpg[['horsepower', 'origin']].apply(add_horse_origin, axis=1)
```

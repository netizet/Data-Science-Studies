```python
import pandas as pd
import numpy as np
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, MaxAbsScaler, RobustScaler, Normalizer, Binarizer
```

```python
df = pd.read_csv('~/Downloads/melb_data.csv')
df
```

```python
df_test = df[['Price', 'Distance', 'Car', 'Landsize', 'Rooms', 'Bedroom2', 'Bathroom']]
df_test.head()
```

# 1. Standard Scaler

```python
df_test['Prize Std Scl'] = StandardScaler().fit_transform(df_test[['Price']])
df_test.head()
```

# 2. Min Max Scaler

```python
df_test['Price MinMax Sc'] = MinMaxScaler().fit_transform(df_test[['Price']])
df_test.head()
```

```python
df_test['Price MaxAbs Sc'] = MaxAbsScaler().fit_transform(df_test[['Price']])
df_test.head()
```

```python
df_test['Price Robust Sc'] = RobustScaler().fit_transform(df_test[['Price']])
df_test.head()
```

```python
df_test['Price Norm'] = Normalizer().fit_transform(df_test[['Price']])
df_test.head()
```
